## Facenet-Pytorch MTCNN Face Detector

```python
#!/usr/bin/env python
# coding=utf-8

import sys
import cv2
import torch
import dlib
import time
import matplotlib.pyplot as plt
import numpy as np

sys.path.insert(1, "MTCNN")
from mtcnn import MTCNN
```

```python
device = 'cuda:0' if torch.cuda.is_available() else 'cpu'
image_path = '1.jpg'
image = plt.imread(image_path)
plt.imshow(image)
```

### Compare Dlib & MTCNN Detection Time

```python
dlib_face_detector = dlib.get_frontal_face_detector()
mtcnn_face_detector = MTCNN(image_size=160, select_largest=False, device=device)

start = time.time()
faces_dlib = dlib_face_detector(image, 1)
end = time.time()
print("Dlib Execution Time:", format(end - start, '.2f'), 's')

start = time.time()
faces_mtcnn, _ = mtcnn_face_detector.detect(image)
end = time.time()
print("Facenet-pytorch Execution Time:", format(end - start, '.2f'), 's')
```

### Draw Bonding Boxes

```python
image = plt.imread(image_path)
for face in faces_dlib:
    cv2.rectangle(image, 
                (face.left(), face.top()), 
                (face.right(), face.bottom()), 
                (0, 255, 0), 2)
    
faces_mtcnn = (np.int64(faces_mtcnn)).reshape(-1, 2, 2)
for face in faces_mtcnn:
    cv2.rectangle(image, tuple(face[0]), tuple(face[1]), (0, 0, 255), 2)
        
img_height, img_width = image.shape[:2]
cv2.putText(image, "Dlib", (img_width-70,30), cv2.FONT_HERSHEY_SIMPLEX, 0.8,
                (0,255,0), 2)
cv2.putText(image, "MTCNN", (img_width-70,60), cv2.FONT_HERSHEY_SIMPLEX, 0.8,
                (0,0,255), 2)

plt.imshow(image, cmap="binary")       
```

### MTCNN Facial Landmark

```python
image = plt.imread(image_path)
faces_mtcnn, _, points = mtcnn_face_detector.detect(image, landmarks=True)
for point in points:
    for pt in point:
        cv2.circle(image, tuple(np.int32(pt)), 1, (255, 255, 255), 12)

plt.imshow(image, cmap="binary")
```
